```python
import pickle
from spread_arb.vbt_backtest import BacktestRunner
from spread_arb.pandas_backtest import DataPipeline
from spread_arb.feeds import Spread, QuotesFeed
from spread_arb.data_prep.fee_info import get_fee_info
import pandas as pd

dp = DataPipeline(
    frequency=pd.Timedelta(minutes=10),
    start_time="2023-04-10",
    end_time="2023-05-10",
)

spreads = dp.run()
fee_info = get_fee_info()
```

```python
quotes_spreads = pd.DataFrame(index=pd.MultiIndex.from_arrays([[], [], []], names=["symbol", "exchange_x", "exchange_y"]))

for idx, spread in spreads.items():
    quotes_a = QuotesFeed(
        spread.get_x().drop(columns=["price_signal", "filled"]),
        exchange_name=idx[1],
        symbol=idx[0],
        inst_type="perpetual",
        interval=dp.frequency,
        starttime=dp.start_time,
        endtime=dp.end_time,
    )
    fee_info_a = fee_info.loc[idx[1], idx[0]].to_dict("records")[0]
    quotes_b = QuotesFeed(
        spread.get_y().drop(columns=["price_signal", "filled"]),
        exchange_name=idx[2],
        symbol=idx[0],
        inst_type="perpetual",
        interval=dp.frequency,
        starttime=dp.start_time,
        endtime=dp.end_time,
    )
    fee_info_b = fee_info.loc[idx[2], idx[0]].to_dict("records")[0]
    spread = Spread.from_quotes(quotes_a, quotes_b)
    quotes_spreads.loc[idx, "spread"] = pickle.dumps(spread)
    quotes_spreads.loc[idx, "fee_info_a"] = pickle.dumps(fee_info_a)
    quotes_spreads.loc[idx, "fee_info_b"] = pickle.dumps(fee_info_b)
```

```python
quotes_spreads
```

```python
bt_runner = BacktestRunner(
    initial_cash=100000,
    trade_value=10000,
    z_score_period=500,
    z_score_thresholds=(0, 1),
    use_slippage=False,
    use_funding_rate=False,
    profitable_only=True,
    log_dir=None,
    force_logging=False,
)
```

```python
res = bt_runner.run_quotes(quotes_spreads)
```
